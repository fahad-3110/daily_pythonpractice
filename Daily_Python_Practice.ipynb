```python
# set methods in python
s1 = {1,2,4,4}
s2 = {1,3,9,4,8}
print(s1,s2)
print(s1.union(s2)) #dono k values ajyngi aur repeat ni hongi agr dono me same b ho to
# s1.update(s2)
print(s1.intersection(s2))  #srf common values ajygi

# symmetric difference
# common values ko minus krdyga
x = {4,5,6,7}
y = {2,3,4,5,9}
print(x.symmetric_difference(y))


#difference and diff update 
# formula (A-B)
a = {4,5,6,7}
b = {2,3,4,5,9}
print(a.difference(y))

#disjoints 
# jin me koi intersection na ho
a = {4,5,6,7}
b = {2,3,4,5,9}
print(a.isdisjoint(y)) 
# yh false is liye aya q k dono me 4 b ha aur 5 b
a = {42,51,6,7}
b = {2,3,4,5,9}
print(a.isdisjoint(y)) 
# yh true is liye ayga q k dono me different value h mtlb k dono me koi common ni hona chyie

# issuperset: kisi b particular set ki value original set me ho to true ayga wrna false
# issubset:  original set ki value particular set me ho to true ayga wrna false

# important method
# ADD()
z = {2,3,4}
z.add(5)
print(z)

# update()

# remove/discard()
# given value remove krdyga
s = {0,1,3,4,9}
s.remove(4)
print(s)
# agr koi value ho he na aur wo remove krny k liye di jye to error ayga
# discard error ni dyga are koi value na ho 
s.discard(4)

# pop()
# set ki last value ko remove krdyga but hmain ni pata k konsa remove kryga
g = {3,5,6}
g.pop()
print(g)

# del()
# del method ni h , 1 keyword h jo entire set ko delete krdyga
j = {3,3,3}
del j

# clear()
# set k saray elements delete krdyga aur srf empty set rh jyga asay set()
k = {3,3,34,5}
k.clear()
print(k)

```

```python
# lambda functions
# A Lambda Function is a one-line, anonymous function in Python. It is used for short operations without defining a full function using def.

# def double(x): yh simple way h
#     return x*2

double = lambda x: x*2 #is sy func define ni krna prta

print(double(10))

add = lambda y: y+10
cube = lambda y: y*y*y
avg = lambda x , y: (x+y)/2
avg = lambda x , y , z: (x+y+z)/2
print(add(10))
print(cube(2))
# print(avg(3,4))
print(avg(3,4,5))
```

```python
# Task: Count Vowels and Consonants
# ðŸ”¹ Objective: Write a Python program that asks the user to enter a word and then counts how many vowels (a, e, i, o, u) and consonants are in the word.

word = input('enter any word').lower()

vowels = 'aeiou'
consonants_count = 0
vowels_count = 0

for letter in word:
    if letter.isalpha(): #agr input koi letter hoto
        if letter in vowels:
            vowels_count += 1
        else:
            consonants_count +=1

print(f'vowels: {vowels_count}')
print(f'consonants: {consonants_count}')

```

```python
# âœ… Objective: Build a basic calculator that can perform addition, subtraction, multiplication, and division.

# ðŸ› ï¸ Steps to Follow:
# 1ï¸âƒ£ Ask the user to enter two numbers.
# 2ï¸âƒ£ Ask the user to choose an operation (+, -, *, /).
# 3ï¸âƒ£ Perform the operation and display the result.
# 4ï¸âƒ£ Use exception handling to prevent errors (e.g., division by zero).
try:
    num1 = int(input('enter 1st number: '))
    num2 = int(input('enter 2nd number: '))
    operation = input('enter these operations: +,-,*,/')
    if operation== '+':
        print('addition of num1 and num2 is:', num1 + num2)
    elif operation == '*':
        print('multiplication of num1 and num2 is:', num1 * num2)
    elif operation == '-':
        print('subtraction of num1 and num2 is:', num1 - num2)
    elif operation == '/':
        print('division of num1 and num 2 is: ' , num1 / num2)
    else:
        print('invalid input')

except ZeroDivisionError:
    print('NUM CANT BE DIVIDED')
except ValueError:
    print('only enter numbers')
```
